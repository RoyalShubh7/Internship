Question 1)  

Scrape the details of most viewed videos on YouTube from Wikipedia.

Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos

You need to find following details:

A) Rank

B) Name

C) Artist

D) Upload date

E) Views


```python
def Shubh(url):   #Creating Function to searches details of most viewed videos on YouTube from Wikipedia
        
    url = "https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos"   #URL of the question.
        
    driver = webdriver.Chrome(r"C:\Driver\chromedriver.exe")  #Connecting to the webdriver.
    driver.get(url)
    time.sleep(3)
# Creating Empty variables to append the particular column data in the variables.

    Rank1 = []
    Name1 = []
    Artist1 = []
    Date1 = []
    Views1 = []

    #Rank
    for i in driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter" ][1]/tbody/tr/td[1]'):
        Rank1.append(i.text)
    #Name
    for i in driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter" ][1]/tbody/tr/td[2]'):
        Name1.append(i.text)
    #artist
    for i in driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter" ][1]/tbody/tr/td[3]'):
        Artist1.append(i.text)
    #Date
    for i in driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter" ][1]/tbody/tr/td[5]'):
        Date1.append(i.text)
    #Views
    for i in driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter" ][1]/tbody/tr/td[4]'):
        Views1.append(i.text)

    #Creating Dataframe
    
    data=pd.DataFrame()
    data['Rank'] = Rank1
    data['Name'] = Name1
    data['Artist'] = Artist1
    data['Date'] = Date1
    data['Views'] = Views1
    
    
    return data  #return the Dataframe With the Collected data.
```

```python
# Calling the Function With Declearing the Variable

df=Shubh('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')

df
```

```python
def fixtures(url):   #Creating Function to searches details of team India’s international fixtures 
    
    Url = "https://www.bcci.tv"     #URL of the question.

    driver = webdriver.Chrome(r"C:\Driver\chromedriver.exe")  #Connecting to the webdriver.
    driver.get(url)
    driver.maximize_window()

    inter=driver.find_element_by_xpath('/html/body/div[3]/div/div[2]/div[2]/nav/ul/li[1]/div[2]')
    fix=driver.find_element_by_xpath('/html/body/div[3]/div/div[2]/div[2]/nav/ul/li[1]/div[2]/div/ul/li[1]/a')
    action=ActionChains(driver)
    action.move_to_element(inter).move_to_element(fix).click().perform()
    time.sleep(2)
    
# Creating Empty variables to append the particular column data in the variables.

    title=[]
    series=[]
    place=[]
    date=[]
    t=[]

    #Title
    for i in driver.find_elements_by_xpath('//strong[@class="fixture__name fixture__name--with-margin"]'):
        title.append(i.text)

    #Series
    for i in driver.find_elements_by_xpath('//span[@class="u-unskewed-text fixture__tournament-label u-truncated"]'):
        series.append(i.text)

    #Place
    for i in driver.find_elements_by_xpath('//p[@class="fixture__additional-info"]/span'):
        place.append(i.text)

    #Date

    day=[]
    month=[]

    #Extracting day
    for i in driver.find_elements_by_xpath('//span[@class="fixture__date"]'):
        day.append(i.text)

    #Extracting month
    for i in driver.find_elements_by_xpath('//span[@class="fixture__month"]'):
        month.append(i.text)

    #joining day and month into one column
    for i in range(len(day)):
        date.append(day[i]+month[i])

    #Time
    for i in driver.find_elements_by_xpath('//span[@class="fixture__time"]'):
        t.append(i.text)


    #Creating Dataframe:
    
    data=pd.DataFrame()
    data['Match Title']=title
    data['Series']=series
    data['Place']=place
    data['Date']=date
    data['Time']=t
    
    
    return data #return the Dataframe With the Collected data.
```

Question 8). Scrape the details of Highest selling novels.

You have to find the following details:
 
 A) Book name
 
 B) Author name
 
 C) Volumes sold
 
 D) Publisher
 
 E) Genre

```python
def novels():       #Creating Function to searches details of Heighest Selling novels.
   
    #URL template for accessing the website.
    
    url = 'https://www.theguardian.com/news/datablog/2012/aug/09/best-selling-books-all-time-fifty-shades-grey-compare/'
    
    driver = webdriver.Chrome(r"C:\Driver\chromedriver.exe")  #Connecting to the webdriver.
    driver.get(url)#Opening with the URL template.
    driver.maximize_window()#Maximize the Window.
    
    #Creating a dummy list and scraping the all the required details.
   
    dummy = [i.text for i in driver.find_elements(By.XPATH,"//table [@class ='in-article sortable']//tr/td")]
    
    driver.close() #Quiting the driver post collecting the details. 
    
    dummy.remove('SOURCE: NIELSEN BOOK SCAN') #Removing the unwanted details from dummy list.
    
    
# Creating Empty variables to append the particular column data in the variables.

    Book_name = [] 
    Author_name = [] 
    Volumes_sold = [] 
    Publisher = [] 
    Genre = [] 

    z = 0
    while z < len(dummy): #Splitting dummy list and storing the required details it in the appropriate list.
        z+=1
        Book_name.append(dummy[z])
        z+=1
        Author_name.append(dummy[z])
        z+=1
        Volumes_sold.append(dummy[z])
        z+=1
        Publisher.append(dummy[z])
        z+=1
        Genre.append(dummy[z])
        z+=1
    
    #Creating the DataFrame with all the collected details.
    table = pd.DataFrame({"Book_name" : Book_name,
                          "Author_name" : Author_name,
                          "Volumes_sold" : Volumes_sold,
                          "Publisher" : Publisher,
                          "Genre": Genre})
    
    return table #returning the DataFrame.   
```

```python
def uci(url):             #Creating Function to searches details of Datasets from UCI machine learning repositories.
    
    
    url = "https://archive.ics.uci.edu"                  #URL of the question.

    driver = webdriver.Chrome(r"C:\Driver\chromedriver.exe")  #Connecting to the webdriver.
    driver.get(url)
    driver.maximize_window()

    #Clicking on show all repositories
    driver.find_element(By.XPATH,'/html/body/table[1]/tbody/tr/td[2]/span[2]/a').click()
    
# Creating Empty variables to append the particular column data in the variables.

    name=[]
    dt=[]
    task=[]
    at=[]
    noi=[]
    noa=[]
    year=[]

    #Name
    for j,i in enumerate(driver.find_elements(By.XPATH,'//table[@border="1"]/tbody/tr/td[1]')):
        if j==0:
            pass
        else:
            name.append(i.text)
    #Data Type
    for j,i in enumerate(driver.find_elements(By.XPATH,'//table[@border="1"]/tbody/tr/td[2]')):
        if j==0:
            pass
        else:
            dt.append(i.text)
    #Task
    for j,i in enumerate(driver.find_elements(By.XPATH,'//table[@border="1"]/tbody/tr/td[3]')):
        if j==0:
            pass
        else:
            task.append(i.text)
    #Attribute Type
    for j,i in enumerate(driver.find_elements(By.XPATH,'//table[@border="1"]/tbody/tr/td[4]')):
        if j==0:
            pass
        else:
            at.append(i.text)
    #No of instances
    for j,i in enumerate(driver.find_elements(By.XPATH,'//table[@border="1"]/tbody/tr/td[5]')):
        if j==0:
            pass
        else:
            noi.append(i.text)
    #No of Attribute
    for j,i in enumerate(driver.find_elements(By.XPATH,'//table[@border="1"]/tbody/tr/td[6]')):
        if j==0:
            pass
        else:
            noa.append(i.text)
    #Year
    for j,i in enumerate(driver.find_elements(By.XPATH,'//table[@border="1"]/tbody/tr/td[7]')):
        if j==0:
            pass
        else:
            year.append(i.text)

    #Creating Dataframe:
    
    data=pd.DataFrame()
    data['Dataset Name']=name
    data['Data type']=dt
    data['Task']=task
    data['Attribute type']=at
    data['No. of instances']=noi
    data['No. of Attribute']=noa
    data['Year']=year
    
    return data        #return the Dataframe With the Collected data.
```
